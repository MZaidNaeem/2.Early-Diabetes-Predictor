<h1>Separating Target and Scaling Whoel Data</h1>

```python
# Separate features and target
X = data.drop('Outcome', axis=1)
y = data['Outcome']

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

<h1>Oversampling minority class using Smoke Tomek</h1>

```python
# Apply SMOTE-Tomek
smote_tomek = SMOTETomek(random_state=42)
X_resampled, y_resampled = smote_tomek.fit_resample(X_scaled, y)

# Show class distribution before and after
print("Original class distribution:", Counter(y))
print("Resampled class distribution:", Counter(y_resampled))
```

<h1>Outlier Detection</h1>

```python
features = data.drop('Outcome', axis=1)

# Plot boxplots
plt.figure(figsize=(16, 12))
for i, column in enumerate(features.columns):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(y=features[column], color='skyblue')
    plt.title(f'Boxplot of {column}')
    plt.tight_layout()

plt.suptitle("Box Plots for Outlier Detection (All Features)", fontsize=18, y=1.02)
plt.show()
```

according to industry knowledge these are usually not outliers these value can exist

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Train model
model = SVC()
model.fit(X_train, y_train)

# Evaluate
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))

accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.4f}")

print("Cross Val Accuray "+str(cross_val_score(model, X_resampled, y_resampled, cv=StratifiedKFold(n_splits=5), scoring='accuracy').mean()))
```

```python
def objective(trial):
    # Define the search space
    C = trial.suggest_float('C', 1e-3, 1e3, log=True)
    kernel = trial.suggest_categorical('kernel', ['linear', 'rbf', 'poly', 'sigmoid'])

    # Additional kernel-specific parameters
    if kernel in ['rbf', 'poly', 'sigmoid']:
        gamma = trial.suggest_float('gamma', 1e-4, 1e1, log=True)
    else:
        gamma = 'auto'

    if kernel == 'poly':
        degree = trial.suggest_int('degree', 2, 5)
    else:
        degree = 3  # default

    # Create model with suggested parameters
    model = SVC(C=C, kernel=kernel, gamma=gamma, degree=degree)

    # Cross-validation
    scores = cross_val_score(model, X_resampled, y_resampled, cv=StratifiedKFold(n_splits=5), scoring='accuracy')
    return scores.mean()


# Run Optuna optimization
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50)

```

```python
# Show best results
print("Best Parameters:", study.best_params)
print("Best Accuracy:", study.best_value)
```

```python
# Train the final model using best hyperparameters
final_model = SVC(
    C=3.3866073719516177,
    kernel='rbf',
    gamma=2.363463904040144
)

# Fit on the full (resampled or original) training data
final_model.fit(X_resampled, y_resampled)

# Now you can use final_model to predict
y_pred = final_model.predict(X_test)  # if X_test is available
```
